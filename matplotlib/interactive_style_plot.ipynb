```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
data = sm.datasets.co2.load_pandas()
co2 = data.data
```

```python
co2.head(3)
```

```python
co2.plot()
plt.show()
```

```python
co2.isnull().sum()
```

```python
co2_nonull = co2.fillna(co2.bfill())
co2_nonull.isnull().sum()
```

```python
from ipywidgets.widgets import Dropdown
from ipywidgets import interact
```

```python
@interact(x=Dropdown(
    options=plt.style.available,
    description='Style:',
    disabled=False))
def plot_graph(x):
    plt.style.use(x)
    co2_nonull.plot(figsize=(15, 6))
    plt.show()
    
```
